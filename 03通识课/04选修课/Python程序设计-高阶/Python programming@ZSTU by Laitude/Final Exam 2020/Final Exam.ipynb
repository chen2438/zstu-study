## 3.回音壁★★★★
回音壁有长距离传输声音效果的原因是皇穹宇围墙的墙面有利于声波的传播。围墙由磨砖对缝砌成，光滑平整，基本符合弹性体波动方程在表面的边界条件，可以很容易地接受到靠近墙面的声波。声波在墙面中传播，也就是在固体中传播，其速度快于在空气中，同时产生的能量损耗也较小，这解释了小声可闻的现象。声波在传播过程中，频率较小的波渗透深度大，损耗也大，有效过滤了低频杂音，同时造成了声音悠长。<br>
这是古代建筑的成就！<br>
<img src='回音壁.jpg' width=180><br>
但现在推荐算法也形成了一个回音壁，就是一旦推荐算法了解到了你的兴趣方向以后，就一直推荐给符合你兴趣的内容，从而会形成一个近似于封闭的内容回音壁，就像声音在回音壁里反射出不去一样。<br>
现在请你验证头条、抖音、快手或其他具有推荐能力的平台，是否也会有这样的回音壁现象。

## 16.社会模型★★★★★
现在，我们来模拟一个社会，在这个社会里，每一个实体都按以下的规则运转：
- 整个社会的总资源每个周期按一个特定的速度在增长，增长的速度与实体的个数相关；
- 每个实体在一个周期中，如果不能摄入资源，那么它就死亡，它所占据的资源就被释放；
- 每个实体在一个周期中，能够摄入它自身占有资源的特定比例；
- 开始的时候每个实体摄入资源的顺序是随机的，但随着周期的增加，一个实体所占据资源越多，那么它摄入资源优先的概率就越大；
- 一个实体摄入资源以后，会产出与自身占据资源特定比例的新资源，但一个实体所占据资源越多，这个比例越低
- 整个社会实体个数稳定以后，模拟结束。

请你选用不同的参数来模拟这个社会模型，看看能够得到什么结论。当然，如果你觉得上面的规则有不合理的地方，你还可以改进它，以更好的模拟现实社会。<br>
实体竞争的过程，最好用可视化的方法演示出来。
