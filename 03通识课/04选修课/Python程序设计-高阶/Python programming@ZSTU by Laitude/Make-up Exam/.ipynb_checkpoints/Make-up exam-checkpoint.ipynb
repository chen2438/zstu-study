## 说明
- 所有补考试题都是开放的，也就是说，并没有标准答案，你的得分完全是根据你<font color=red>**最终的方法、努力的过程、提交的程序和报告**</font>来决定。

- 如果对应的题目有数据文件，则以题号相同名称的文件都为该题的数据文件，例：4.png，4.rar都是4号题的数据文件。
- <font color=red>**每一题只能一个小组做，所以注册前，请务必确认该题没有被占用，并截图发到Python通知群**。</font>
- <font color=red>**注册必须以小组注册，每组3人，否则注册无效**。</font>

**答辩事项：**
1. 答辩顺序为出题顺序；
2. 答辩时间在下学期开始的另行通知；
3. 答辩人按如下规则选定：将参与度从小到大排序，如果两人的参与度之差大于5%，则在二人之间将整个小组分段，答辩人在参与度最少的段内随机选取；如果参与度均小于5%，则每人都对自己的内容进行答辩；
4. 提交内容：notebook文件和pdf文件（notebook文件的打印稿，评分表excel文件为首页和第2页）各一份，包括：
  - 必备信息：题目，团队名称，团队组成人员名单（包含学号）；
  - 技术思路和解决方案（包含代码）；
 - 运行结果；
 - 自我评价，哪些地方做得好，哪些不行，要改进。
 - 文件内容要求图文并茂。
5. 成绩评定：以最终答辩和提交的报告作为评价依据，给答辩人评分，<font color=red>**小组总分为选题成绩乘以小组人数，各成员的分数参与度进行分配，教师有权利根据实际情况进行调整**</font>。

# 2019补考题

## 1.医疗问题★★★
最近的民航总医院行杨医生离世让医患关系又成为热点。<br>
诚然，目前的医疗状况的确不能让很多人满意，但另一方面，医生们的超负荷工作也是事实。<br>
我们必须对此有一个客观的认识。<br>
请你搜索相关数据，可以包括：医疗费用，治疗时间，治愈率等等，客观地来分析和对比我国的医疗水平与世界水平，然后给出你的结论。

## 2.高考改革★★★★
浙江对高考进行了改革，物理、化学、技术、生物等成为选考科目，允许考2次，成绩在2年内有效。<br>
为了让不同次的考试成绩可以相互比较，成绩是按名次进行赋分。赋分的方式如下图。<br>
<img src='1.jpg' width=320><br>
但是，现实出现了较多问题，比如物理选考人数太少，而选考的都是物理好的，会导致好生得不到相对好的成绩，如果试卷区分度不够，很容易出现某一个分数段扎堆的情况。<br>
现在，假定某一门课考试有6.5万人，但是，由于试卷区分度不太好，可能某一个分数段人很多，造成赋分不是很合理的情况。<br>
请你设计一个方案，能否缓解甚至解决这一赋分问题。<br>
要求：
1.  所有数据由你来生成，可以参考前几年的真实数据；
2.  你的方案的合理性请进行充分的分析。

# 2020补考题

## 1. 国家金字塔
国家与国家之间，也是有层级的，这种层级，可以通过不同的指标来进行划分。<br>
现在，请你选择几个特定的指标，对地球上的国家进行分级，并用可视化的方法，把这种层级表示出来。<br>
有了这个层级，请你再回答一下：在这个层级中，各个国家为什么处在这个层级中，有没有可能出现层级的越迁和掉落？特别是对中国而言。

## 2. 美元与房价
现在，拜登上台，可以马上又要释放2万亿，这些美元又会流向全世界，稀释了美国的通货膨胀，推动了其他国家的通货膨胀。<br>
正因为有了这样的情况，据说刚发布的消息称全国30个城市房价同比都增长了。<br>
现在，请你分析一下，美元周期（即美联储加息和减息的周期）对我国的房价有怎样的影响？包括过去和未来？

## 3.信用
什么是信用？信用就是：欠钱不要紧，只要会还钱，而且是按约定的还，欠得越多，还得越多，信用越好。好的信用，小到可以开个小店，大到可以战争建国。<br>
信用背面就是老赖。现在，请你找相关的数据，看看我国的老赖们是不是变多了？在哪里？是不是随着国家信用体系的建设，越来越不好混了？<br>
可视化分析，肯定是必不可少的。
