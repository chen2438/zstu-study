```python
s = 'Hello world\n文本文件的读取方法\n文本文件的写入方法\n'
f=open('sample.txt', 'w')    #默认使用cp936编码
f.write(s)
f.close()
f=open('sample.txt')         #默认使用cp936编码
print(f.read())
f.close()
```

```python
!type sample.txt
```

```python
f=open('sample.txt')         #默认使用cp936编码,注意mode这里没有给出
f.seek(15)
print(f.tell())    # 注意位置返回以字节为单位
print(f.read(6))    # 字符数!!!
print(f.tell())    # 注意位置返回以字节为单位
f.close()
```

```python
f=open('sample.txt','rb')
f.seek(10)
print(f.tell())
a=f.read(11)    # a是一个字节串
# print(a)
# print(str(a))    # 直接转换,默认的文件编码是utf-8
print(str(a, encoding='cp936'))    # 按特定编码转换成字符串
print(f.tell())
f.seek(0)
a=f.read(200)
print(a)
print(len(a))    # 读取的长度小于要读取的字节数,说明到了文件末尾
f.close()
```

```python
# 一个演示读取文本文件全部内容的代码
f=open('sample.txt','r')
line=f.readline()
while line:    # 注意这个确认是否文件结束的判断,因为读到一个空行,说明文件结束
    print(line, end='\0')
    line=f.readline()
f.close()
```

### 遍历路径

```python
import os 

def listdir(path):
    if not os.path.isdir(path):
        print('Error:"',path,'" is not a directory or does not exist.')
        return
    list_dirs = os.walk(path)                #os.walk返回一个元组，包括3个元素：
                                                #所有路径名、所有目录列表与文件列表
    for root, dirs, files in list_dirs:       #遍历该元组的目录和文件信息
        for d in dirs: 
            print(os.path.join(root, d))     #获取完整路径
        for f in files: 
            print(os.path.join(root, f))     #获取文件绝对路径

listdir('d:\\')
```
