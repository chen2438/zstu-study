# 2. 为什么我不用分期★★★
来自一个真实的例子。<br>
我要还银行3395.31元，然后分成6个月还，分期手续费率是0.57%，分期手续费每月19.35元，分6期还清。<br>
现在请问，按这样的方式还，实际的费率折算成年利率是多少？然后回复你愿意分期吗？
提示：贷款按揭的计算公式如下<br>
$P_t=PM^t-Y(\frac{M^t-1}{M-1})$，其中$M=1+\frac{I}{12}$，$I$是年利率，$t$是按揭的期数，如果按揭3年，就是36期，<br>
$Y$是每个月付的钱（<font color=red>注意每个月还的钱是一样的</font>），$P$是你分期的金额，$P_t$是你第$t$个月后，还剩下要还的钱。<br>
<font color=red>这道题一定要搞清楚的是，分期+手续费，事实上就是按揭还款。</font>
