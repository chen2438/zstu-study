```python
import math

loan=3395.31
rate=0.57/100
step=6
service=loan*rate  # 月手续费
month_fee=service+loan/step
M=1.01
count=0
while True:
    sub=loan*(M**step+1)-month_fee*(M**step-1)/(M-1)
    print(sub, M)
    if sub>0:
        M-=0.000001
    elif sub<0:
        M+=0.000001
    if 0<3395.31-sub<0.01:
        break
    count+=1
    if count>1000:
        break
print((M-1)*12)
```

提示：贷款按揭的计算公式如下<br>
$P_t=PM^t-Y(\frac{M^t-1}{M-1})$，其中$M=1+\frac{I}{12}$，$I$是年利率，$t$是按揭的期数，如果按揭3年，就是36期，<br>
$Y$是每个月付的钱（<font color=red>注意每个月还的钱是一样的</font>），$P$是你分期的金额，$P_t$是你第$t$个月后，还剩下要还的钱。<br>
