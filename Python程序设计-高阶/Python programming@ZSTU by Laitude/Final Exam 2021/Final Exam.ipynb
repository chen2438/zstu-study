## 2.智能小车规划★★★★★
<img src='agv.png' width=1024><br>
上图是一个智能小车库房场地的布局图，其中，右边蓝色的点是装载或卸货的位置，称为port，绿色的点是小车充电的位置橙色的点是货架的位置（有些带了方框，有些没有带方框），深棕色的点是道路的点。<br>
以下是一些假定：
- 小车必须从道路点到达货架位置，然后将货架装上小车，这个过程由小车自动完成，不消耗时间
- 出入库必须用小车将货架装载运送到port，货物装上或卸下后，小车再将货架送到原来的位置
- 每个货架能够存放的货物的总数是一样的，比如都是20件
- 小车每走过特定的距离，就要充电，去充电时走的路不算在这个距离内
- 每一件货的装载和卸货的时间是一样的，由你指定
- 小车转弯需要消耗时间，消耗的时间可以和从一个道路点走到另一个道路点的时间相同

下面，请你用可视化的方式模拟你规划出入库的情况，要求：
- 不同的状态的小车，使用不同颜色的环来显示
- 要考虑小车对道路的占用，也就是说，如果有一个小车停在路上，其他小车必须要绕路
- 要考虑小车对port的占用，也就是说如果有一个小车在port上，其他小车必须等待
- 小车有随机损坏的可能，如果损坏了，就停在那里，停的时间随机生成，这个时间一过，小车可以自动恢复

我们希望这些小车有最大的出入库量，在模拟出入库的同时，显示当前有多少个小车，出库或入库了多少件货

## 16.社会模型★★★★★
现在，我们来模拟一个社会，在这个社会里，每一个实体都按以下的规则运转：
- 整个社会的总资源每个周期按一个特定的速度在增长，增长的速度与实体的个数相关；
- 每个实体在一个周期中，如果不能摄入资源，那么它就死亡，它所占据的资源就被释放；
- 每个实体在一个周期中，能够摄入它自身占有资源的特定比例；
- 开始的时候每个实体摄入资源的顺序是随机的，但随着周期的增加，一个实体所占据资源越多，那么它摄入资源优先的概率就越大；
- 一个实体摄入资源以后，会产出与自身占据资源特定比例的新资源，但一个实体所占据资源越多，这个比例越低
- 整个社会实体个数稳定以后，模拟结束。

请你选用不同的参数来模拟这个社会模型，看看能够得到什么结论。当然，如果你觉得上面的规则有不合理的地方，你还可以改进它，以更好的模拟现实社会。<br>
实体竞争的过程，最好用可视化的方法演示出来。
