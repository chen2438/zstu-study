#### 括号后面的冒号必不可少
#### 函数体相对于def关键字必须保持空格或TAB缩进

```python
def harmonic(n):    # 函数名，参数
    total = 0.0    #函数体，注意缩进
    for i in range(1,n+1):
        total += 1.0 / i
    return total

harmonic(100000)    # 函数调用，注意传递了一个参数100
```

### 形参、实参与返回值
#### 函数<font color=red>形参不需要声明其类型</font>，也<font color=green>不需要指定函数返回值类型</font>
#### 即使该函数不需要接收任何参数，也必须保留一对空的圆括号
#### 允许有多个形参，之间用逗号分隔
#### 函数调用时向其传递实参，将实参值传递给形参

```python
def fib(n):
    a, b = 0, 1
    while a < n:
        print(a, end=' ')
        a, b = b, a+b
    print()    # 没有return语句

fib(1000)
```

```python
def demo(newitem,old_list=[]):
    old_list.append(newitem)
    print(id(old_list))
    return old_list
print(demo('5',[1,2,3,4]))   #right
print(demo('aaa',['a','b'])) #right
print(demo('a'))             #right
print(demo('b'))             #掉坑里了，参数要协调
```

### 变量作用域
#### 全局变量和局部变量
#### 函数外部定义和函数内部定义的全局变量，关键字global
#### 局部变量与全局变量同名，在重合的作用域，局部变量使用优先

```python
x = 5    # 函数外部定义
def demo():
    global x    # 使用全局变量x
    x = 3
    y = 4       # 定义变量，默认为局部变量
    print(x,y)

print(x)
demo()
print(x)
```

### lambda表达式
#### 声明匿名函数，也就是没有函数名字的临时使用的小函数，尤其适合需要一个函数作为另一个函数参数的场合
#### 该表达式的计算结果可以看作是函数的返回值，不允许包含其他复杂的语句，但在表达式中可以调用其他函数
#### lambda 参数:表达式

```python
f = lambda x, y, z: x+y+z        #可以给lambda表达式起名字
f(1,2,3),f('a','b','c')
```

```python
g = lambda x, y=2, z=3: x+y+z    #参数默认值
g(1)
```

```python
g(2, z=4, y=5)                #关键参数
```

```python
def demo(n):
    return n**2

a_list = [1,2,3,4,5]
f=lambda x: demo(x) # x**2
f(5)
print(map(lambda x: demo(x), a_list))
list(map(lambda x: demo(x), a_list))
list(map(demo,a_list))
#在lambda表达式中调用函数,相当于list(map(demo,a_list))
# map() 会根据提供的函数对指定序列做映射。
# 第一个参数 function 以参数序列中的每一个元素调用 function 函数，返回包含每次 function 函数返回值的迭代器。
```

```python
foo = [2, 18, 9, 22, 17, 24, 8, 12, 27]
print(list(filter(lambda x: x % 3 == 0, foo)))
print(list(map(lambda x: x * 2 + 10, foo)))
```

### @装饰器
python的装饰器本质上是一个Python函数，<br>
它可以让其他函数在不需要做任何代码变动的前提下增加额外功能，<br>
装饰器的返回值也是一个函数对象。简单的说装饰器就是一个函数来修改另一个函数。

```python
def hi():
    print('I love ASTM')
 
hi()
```

```python
def hi():
    print('I love ASTM')
    print('I love ISO, too')  # 增加了功能，在函数体中增加代码
    
hi()
```

```python
def hi(ext):
    print('I love ASTM')
    print('I love ISO, too')
    print(ext) #增加了功能，通过使用参数
    
hi('I love IEEE, too')
```

```python
def ext():
    print('I love ISO, too')
    
def hi(a):
    print('I love ASTM')
    a() # 调用参数，参数是一个函数，通过参数是一个函数来增加功能
    
hi(ext)   # 函数作为参数
```

```python
def hi(a):
    print('I love ASTM')
    a()

@hi   # 类似于js里的把整个函数作为参数调用
def ext():
    print('I love ISO, too, in ext')
    print('I hate IEEE, in ext')
    print('I hate ACM, in ext')


# hi(func()
#{
#   print('I love ISO, too')
#   print('I hate IEEE')
#   print('I hate ACM')
#})
#
```

```python
def hi(a):
    print('I love ASTM')
    a()

def ext():
    print('I love ISO, too, in ext')
    print('I hate IEEE, in ext')
    print('I hate ACM, in ext')
    
hi(ext)
```

```python
def hi(a):
    print('I love ASTM in hi')
    return a

@hi   # 类似于js里的把整个函数作为参数调用，大家看看效果
def ext():
    print('I love ISO, too, in ext')
    print('I hate IEEE, in ext')
    print('I hate ACM, in ext')

ext() # hi(ext)
```

#### 实质上装饰器就是用函数hi增加了函数ext的功能

#### 在不修改index函数的源代码以及调用方式的前提下为其添加统计运行时间的功能

```python
import time
def index(x,y):
    time.sleep(3)
    print('index %s %s' % (x,y))
index(11,12)
```

```python
def timer(func):
    def wrapper(*args, **kwargs):
        print('now in wrapper in timer')
        start=time.time()
        res = func(*args, **kwargs)
        stop = time.time()
        print(stop - start)
        return res
    return wrapper

@timer    # timer(index)
def index(x,y,z):
    time.sleep(3)
    print('index %s %s %s in index' % (x,y,z))
    
index(x=11,y=12,z=13)  # 用timer增加了功能以后的index函数
```
