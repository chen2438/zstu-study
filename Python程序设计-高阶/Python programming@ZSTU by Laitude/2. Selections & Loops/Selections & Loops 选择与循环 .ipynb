## Python条件表达式一个约定
### 在选择和循环结构中，条件表达式的值只要<font color=#FF0000>不是False、0（或0.0、0j等）、空值None、空列表、空元组、空集合、空字典、空字符串、空range对象或其他空迭代对象</font>，Python解释器均认为与True等价。
### 简单来说就是<font color=#FF0000>简单类型值如果不是False或0</font>，<font color=#FF0000>复杂类型值不是空</font>，那么就是True。
### 所以，几乎所有的表达式都可以作为条件表达式。

### 多分支结构
### Python没有switch语句，只能用if语句的组合来实现多分支结构

```python
score = 50
if score>100:
    print('wrong score.must<=100.')
else:
    if score>=90:
        print('A')
    else:
        if score>=80:
            print('B')
        else:
            if score>=70:
                print('C')
            else:
                if score>=60:
                    print('D')
                else:
                    if score>=0:
                        print('E')
                    else:
                        print('wrong score.must>0')
```

```python
# 缩写形式
score = 50
if score>100:
    print('wrongscore.must<=100.')
elif score>=90:
    print('A')
elif score>=80:
    print('B')
elif score>=70:
    print('C')
elif score>=60:
    print('D')
elif score>=0:
    print('E')
else:
    print('wrongscore.must>0')
```

```python
# 另外一个例子
x, y = 10,-1
if x>0:
    if y>0:
        print('in first quadrant')
    else:
        print('in fouth quadrant')
else:
    if y>0:
        print('in second quadrant')
    else:
        print('in third quadrant')
```

## 循环结构

### while 循环
while循环一般用于<font color=red>难以提前确定</font>的情况，也可以用于<font color=blue>循环次数确定</font>的情况。<br>
<center><img src='2.png'></center>

```python
# 求1加到100的和
sum = 0
i = 1
while i <= 100:
    sum += i
    i += 1
print("sum is", sum) # sum is 5050
```

### for 循环
for循环一般用于循环次数可以<font color=red>提前确定</font>的情况，注意这种提前确定是通过<font color=red>枚举序列或迭代对象</font>中的元素来实现的。<br>
<center><img src='3.png'></center>

```python
for i in range(10,1,-1):  # 注意range函数的三个参数，分别是起点、终点（不包含）和步长
    print(i)
```

```python
100 in range(101), 101 in range(101) # range(0,101,1)
```

```python
sum = 0
for i in range(101):  # for (i=0;i<101;i++)
    sum +=i
print("sum is", sum) # sum is 5050
```

### break 和 continue

```python
import math
n=100
k=math.sqrt(n)
int(k)
```

```python
#判断素数
import math
n = input('please input a number larger than 1:')
n = int(n)
k = int(math.sqrt(n))  # 调用math包中的sqrt函数求平方根
for i in range(2,k+1):
	if n % i ==0:
		print(n, 'is not a prime number')
		break
else:
    print(n, 'is a prime number!')
```

```python
# 输出100以内的奇数
for i in range(0,100,1):
    if i % 2 ==0:
        continue
    print(i, '\t', end='')    # print不换行的写法,注意这个end,end就表示print将如何结束，默认为end="\n"（换行）
```

### 循环嵌套（多重循环）
#### 循环内部再出现循环
#### break和continue只能对所在循环起作用

```python
for i in range(2,2):
    print(i)
```

```python
# 求100以内的素数
import math
for i in range(2, 100):
    k = int(math.sqrt(i))
    for j in range(2, k+1):
        if i % j ==0:
            break
    else:
        print(i,'\t',end='')
```

#### 一个求$\pi$的例子（蒙特卡洛）
蒙特卡洛方法是使用随机数和概率来解决问题，在计算机数学、物理、化学和经济方面都有非常广泛的应用。

<center><img src='Pi_30K.gif'></center>

```python
from random import random
from math import sqrt
darts=2**25
hist=0
for i in range(1,darts):
    x,y=random(),random()
    dist=sqrt(x**2+y**2)
    if dist<=1.0:
        hist=hist+1
pi=4*(hist/darts)
print('pi is %s'%pi)
```

- 假定你借了10万，借期1年，利息是3%，现在请你计算：
- 如果是按揭，每个月还多少？
- 根据按揭实际还款额，折算成先息后本方式，年利率是多少？

贷款按揭的计算公式如下<br>
$P_t=PM^t-Y(\frac{M^t-1}{M-1})$，其中$M=1+\frac{I}{12}$，$I$是年利率，$t$是按揭的期数，<br>
如果按揭3年，就是36期，<br>
$Y$是每个月付的钱（<font color=red>注意每个月还的钱是一样的</font>），$P$是你分期的金额，也就是借的总金额，<br>
$P_t$是你第$t$个月后，还剩下要还的本金。<br>
